# Bokeh

```python
# Q.01] Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x. ?

import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

x = np.linspace(0, 10, 100)
y = np.sin(x)

p = bokeh.plotting.figure(title = 'Sine Wave', x_axis_label = 'x', y_axis_label = 'sin(x)')
p.line(x, y)
bokeh.plotting.show(p)
```

```python
# Q.02] Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns ?
from bokeh.plotting import figure, show
np.random.seed(42)
x = np.random.rand(50) * 100
y = np.random.rand(50) * 100
sizes = np.random.randint(5, 30, 50)
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], 50)

p = figure(title="Bokeh Scatter Plot",x_axis_label='X Value', y_axis_label='Y Value')
p.scatter(x, y, size=sizes, color=colors, alpha=0.6)
show(p)
```

```python
# Q.03] Generate a Bokeh bar chart representing the counts of different fruits using the following dataset. ?
'''fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]'''


fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count')
p.vbar(x=fruits, top=counts, width=0.5, color="orange")
show(p)
```

```python
# Q.04] Create a Bokeh histogram to visualize the distribution of the given data. ?
'''data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)'''

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p = figure(title='Histogram', x_axis_label='Value', y_axis_label='Frequency')
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='skyblue', line_color='white')
show(p)
```

```python
# Q.05]  Create a Bokeh heatmap using the provided dataset ?
''' data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)'''

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.palettes import Viridis256

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

x_flat = xx.flatten()
y_flat = yy.flatten()
values = data_heatmap.flatten()

color_indices = (values * 255).astype(int)
colors = [Viridis256[i] for i in color_indices]

p = figure(title="Simple Heatmap", x_range=(0, 1), y_range=(0, 1),tools="", toolbar_location=None)

p.rect(x=x_flat, y=y_flat, width=0.1, height=0.1, color=colors, line_color=None)
show(p)
```
